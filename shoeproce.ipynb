```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data.csv")
#Show the first 15 contents
df
```

```python
#Shape of dataset
df.shape
```

```python
#Columns of dataset
df.columns
```

```python
#Dataframe summary 
def summary(df):
    print(f'data shape: {df.shape}')
    summ = pd.DataFrame(df.dtypes, columns=['Data Type'])
    summ['Missing#'] = df.isna().sum()
    summ['Missing%'] = (df.isna().sum())/len(df)
    summ['Dups'] = df.duplicated().sum()
    summ['Uniques'] = df.nunique().values
    summ['Count'] = df.count().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summ['Min'] = desc['min'].values
    summ['Max'] = desc['max'].values
    summ['Average'] = desc['mean'].values
    summ['Standard Deviation'] = desc['std'].values
    summ['First Value'] = df.loc[0].values
    summ['Second Value'] = df.loc[1].values
    summ['Third Value'] = df.loc[2].values

    display(summ)

summary(df)

#There are duplicate of the same shoe brand with same size
```

```python
#Delete duplicate
df = df.drop_duplicates()
```

```python
# Highest Shoes according to Price
df.sort_values(["price"], axis=0, ascending=False, inplace=True)
df.head(10)
```

```python
# Group by 'color' and find the maximum price for each color
color_max_price = df.groupby('color')['price'].max()

# Plotting the graph
plt.figure(figsize=(10, 6))
color_max_price.sort_values(ascending=False).plot(kind='bar', color='skyblue')

# Adding titles and labels
plt.title('Highest Price of Shoes by Color', fontsize=14)
plt.xlabel('Color', fontsize=12)
plt.ylabel('Highest Price', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

```

```python
# Group by 'size' and find the maximum price for each size
size_max_price = df.groupby('size')['price'].max()

# Plotting the graph
plt.figure(figsize=(10, 6))
size_max_price.sort_values(ascending=False).plot(kind='bar', color='lightgreen')

# Adding titles and labels
plt.title('Highest Price of Shoes by Size', fontsize=14)
plt.xlabel('Size', fontsize=12)
plt.ylabel('Highest Price', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by 'brand' and find the average price for each brand
brand_avg_price = df.groupby('brand')['price'].mean()

# Plotting the graph
plt.figure(figsize=(12, 6))
brand_avg_price.sort_values(ascending=False).plot(kind='bar', color='lightcoral')

# Adding titles and labels
plt.title('Average Price of Shoes by Brand', fontsize=14)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by 'brand' and find the maximum price for each brand
brand_max_price = df.groupby('brand')['price'].max()

# Plotting the graph
plt.figure(figsize=(12, 6))
brand_max_price.sort_values(ascending=False).plot(kind='bar', color='lightgreen')

# Adding titles and labels
plt.title('Maximum Price of Shoes by Brand', fontsize=14)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Maximum Price', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
